```python
data = [{'group':x.split('\t')[0], 'category':x.split('\t')[1], 'group_name':x.split('\t')[2]} for x in hs.split('\n')]
data
```

```python
df = pd.DataFrame(data)
df
```

```python
df.to_csv('hs_code.csv', index=False)
```

```python
datas = []
for d in data:
    lower, upper = [int(r) for r in d['category'].replace(' ','').split('–')]
    for i in range(lower, upper+1):
        datas.append({
            'hs_code':"{:04d}".format(i),
            'name':d['group_name']
        })
datas
```

```python
df2 = pd.DataFrame(datas)
df2
```

```python
df2.to_csv('hs_code_split.csv', index=False)
```
